```python
import cv2                
import matplotlib.pyplot as plt                        
%matplotlib inline  

from sklearn.datasets import load_files       
from keras.utils import np_utils
import numpy as np
from glob import glob

from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input as preprocess_input_resnet50
```

## Data Loading

The data is divided into three directories:
- 'train' directory contains 6679 images. This directory will be used as a training set to train our model.
- 'valid' directory contains 835 images. This directory will be used as a validation set to evaluate our model.
- 'test' directory contains 836 images. This directory will be used to test our model.

The data is from a kaggle challenge, so it can be downloaded from the official page of the challenge from [here](https://www.kaggle.com/c/dog-breed-identification). Or you can download the data directly from [here](https://s3-us-west-1.amazonaws.com/udacity-aind/dog-project/dogImages.zip). [1.0 GB]

```python
def load_dataset(path):
    """
    This function takes the data path on your computer.
    Then it loads the images and the targets and return each as a tuple.
    """
    data = load_files(path)
    dog_files = np.array(data['filenames'])
    dog_targets = np_utils.to_categorical(np.array(data['target']), 133)
    return dog_files, dog_targets

train_files, train_targets = load_dataset('dogImages/train')
valid_files, valid_targets = load_dataset('dogImages/valid')
test_files, test_targets = load_dataset('dogImages/test')

dog_names = [item[20:-1] for item in sorted(glob("dogImages/train/*/"))]

# Let's check the dataset
print('There are %d total dog categories.' % len(dog_names))
print('There are %s total dog images.\n' % len(np.hstack([train_files, valid_files, test_files])))
print('There are %d training dog images.' % len(train_files))
print('There are %d validation dog images.' % len(valid_files))
print('There are %d test dog images.'% len(test_files))
```

## Features Extraction

This part is the most important part. Here, we turn the raw images into numbers which can be used as features. The given images had some problems:
- They're high resolution images.
- They differ in size.
- They are raw and un-engineered.

In the following few functions, we will handle these problems

```python
from keras.preprocessing import image                  
from tqdm import tqdm

def path_to_tensor(img_path):
    """
    This function takes a path to one image as an input and does the following:
    -> resize the image to 224 x 224 x 3 pixels (first line). 
       The first 224 refers to the image width, the second 224 refers to the height and 3 is the number of channels (RGB)
       knowing that R is read, G is green and B is blue.
    -> convert the image into features/tensor (second line)
    And it returns the tensor as 4D matrix instead of 3D which could make things easier afterward.
    So, if the tensor shape is (224 x 224 x 3), it will be returned as (1 x 224 x 224 x 3)     for example, """
    # loads RGB image as PIL.Image.Image type
    img = image.load_img(img_path, target_size=(224, 224))
    # convert PIL.Image.Image type to 3D tensor with shape (224, 224, 3)
    x = image.img_to_array(img)
    # convert 3D tensor to 4D tensor with shape (1, 224, 224, 3) and return 4D tensor
    return np.expand_dims(x, axis=0)




def paths_to_tensor(img_paths):
    """
    This function applies the previous function (path_to_tensor) over all the images
    """
    list_of_tensors = [path_to_tensor(img_path) for img_path in tqdm(img_paths)]
    return np.vstack(list_of_tensors)
```

```python
def extract_Resnet50(file_paths):
    """
    This function is the most important function as it takes data path and does the following:
    -> First, apply the (paths_to_tensor) function over all the images to turn them into featuers.
    -> Reshape these features to be applicable for the ResNet50 model.
    -> Run the ResNet50 model over the engineered features.
    return the prediction of the ResNet50 model. Its shape should be (no.data, 7, 7, 2084)
    """
    tensors = paths_to_tensor(file_paths).astype('float32')
    print('Normalizing: Done!')
    preprocessed_input = preprocess_input_resnet50(tensors)
    print('Preprocessing the input: Done!')
    #batch_size is a very important paramter that you should look up. Try Googling 'BGD' or 'Batch Gradient Descent'
    res= ResNet50(weights='imagenet', include_top=False).predict(preprocessed_input, batch_size=32)
    return res
```

```python
#apply extract_Resnet50 over the training directory
train_resnet50 = extract_Resnet50(train_files)
```

```python
#apply extract_Resnet50 over the validation directory
valid_resnet50 = extract_Resnet50(valid_files)
```

```python
#apply extract_Resnet50 over the test directory
test_resnet50 = extract_Resnet50(test_files)
```

## Model Preperation

Now, we have applied ResNet50 over the dog breed dataset, and it got (7 x 7 x 2048) matrix as features for every image. Now, we are going to create a convolution neural nework model that can use these matrices as input and return the dog breed.

```python
from keras.layers.pooling import GlobalAveragePooling2D
from keras.layers.merge import Concatenate
from keras.layers import Input, Dense
from keras.layers.core import Dropout, Activation
from keras.callbacks import ModelCheckpoint
from keras.layers.normalization import BatchNormalization
from keras.models import Model

def input_branch(input_shape=None):
    """
    This function creates the input layer of the new CNN model
    """
    size = int(input_shape[2] / 4)
    branch_input = Input(shape=input_shape)
    #apply average pool over the input
    branch = GlobalAveragePooling2D()(branch_input)
    #connect the input layer with the weights and apply the activation function over (weights*input+bias)
    branch = Dense(size, use_bias=False, kernel_initializer='uniform')(branch)
    #apply BatchNormalizatin technique to provide overfitting
    branch = BatchNormalization()(branch)
    #use relu (Rectified Linear Unit) as activation function
    branch = Activation("relu")(branch)
    return branch, branch_input

# create input layer using input_branch function
resnet50_branch, resnet50_input = input_branch(input_shape=(7, 7, 2048))
# use 30% dropout, which is a way to prevent overfitting. You can google 'droupout technique'
net = Dropout(0.3)(resnet50_branch)
#connect the input layer with the weights and apply the activation function over (weights*input+bias)
net = Dense(640, use_bias=False, kernel_initializer='uniform')(net)
#apply BatchNormalizatin technique to provide overfitting
net = BatchNormalization()(net)
#use relu (Rectified Linear Unit) as activation function
net = Activation("relu")(net)
# use 30% dropout
net = Dropout(0.3)(net)
net = Dense(133, kernel_initializer='uniform', activation="softmax")(net)

#connect all these layers into one model
model = Model(inputs=[resnet50_input], outputs=[net])
model.summary()
```

## Model Training

```python
#compile is used to hard-core connect the layers together and initialize the necessary parameters
model.compile(loss='categorical_crossentropy', optimizer="rmsprop", metrics=['accuracy'])
#this object is used to save the model
checkpointer = ModelCheckpoint(filepath='saved_models/bestmodel.hdf5', 
                               verbose=1, save_best_only=True)

#apply the new model over the data now.
model.fit([train_resnet50], train_targets, 
          validation_data=([valid_resnet50], valid_targets),
          epochs=10, batch_size=4, callbacks=[checkpointer], verbose=1)
```

## Model Testing

In this part, we are going to see how our model is good at recognizing dog breeds. We are going to best model according to the validation score. Then, use that model to predict the dog breeds given dog images. Finally, we get the accuracy of that model.

Let's do all of that

```python
#load the model that has least validation error
model.load_weights('saved_models/bestmodel.hdf5')
```

```python
from sklearn.metrics import accuracy_score

#use the model to predict
predictions = model.predict([test_resnet50])
breed_predictions = [np.argmax(prediction) for prediction in predictions]
breed_true_labels = [np.argmax(true_label) for true_label in test_targets]
#evaluate the model
print('Test accuracy: %.4f%%' % (accuracy_score(breed_true_labels, breed_predictions) * 100))
```

```python
def extract_image_Resnet50(file_paths):
    """
    Preprocess just one image
    """
    tensors = path_to_tensor(file_paths).astype('float32')
    print('normalizing done')
    preprocessed_input = preprocess_input_resnet50(tensors)
    print('preprocessed input done')
    res= ResNet50(weights='imagenet', include_top=False).predict(preprocessed_input, batch_size=32)
    print(res.shape)
    return res
```

```python
def dog_breed(filePath):
    features=extract_image_Resnet50(filePath)
    pred=model.predict(features)
    return dog_names[np.argmax(pred)]
```

```python
def dog_breed_predictor(img_path):
    # determine the predicted dog breed
    breed = dog_breed(img_path) 
    # display the image
    img = cv2.imread(img_path)
    cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(cv_rgb)
    plt.show()
    # display relevant predictor result
    print(breed)
```
